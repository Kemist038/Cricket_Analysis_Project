## Data Cleaning & Manipulating part

```python
cricket_data
```

```python
cricket_data.info()
```

```python
cricket_data["Player"].unique()
```

```python
cricket_data["Country"] = cricket_data["Player"].str.split("(").str.get(1).str.split(")").str.get(0)
```

```python
cricket_data.head()
```

```python
cricket_data["Country"].unique()
```

```python
cricket_data["Country"].replace("ICC/","", regex = True, inplace = True)
```

```python
cricket_data["Country"].unique()
```

```python
cricket_data["Player"] = cricket_data["Player"].str.split("(").str.get(0).str.strip()
```

```python
cricket_data.head()
```

```python
cricket_data["Career_Length"] = (cricket_data["Span"].str.split("-").str.get(1).astype(int)) - (cricket_data["Span"].str.split("-").str.get(0).astype(int))
```

```python
cricket_data["Current/End_Year"] = cricket_data["Span"].str.split("-").str.get(1).astype(int)
```

```python
cricket_data["Debut_Year"] = cricket_data["Span"].str.split("-").str.get(0).astype(int)
```

```python
cricket_data.drop("Span", axis =1, inplace = True)
```

```python
cricket_data.head()
```

```python
cricket_data.columns = ['Player', 'Matches', 'Innings', 'NotOut', 'Runs', 'High_Score', 'Average_Rund', 'Balls_Faced', 'Strike_Rate', '100s',
       '50s', '0s', '4s', '6s', 'Country', 'Career_Length', 'Current/End_Year',
       'Debut_Year']
```

```python
cricket_data = cricket_data.reindex(columns = ['Country', 'Player', 'Career_Length', 'Debut_Year', 'Current/End_Year', 'Matches',
                                'Innings', 'NotOut', 'Runs', 'High_Score', 'Average_Rund', 'Balls_Faced', 'Strike_Rate',
                                '100s', '50s', '0s', '4s', '6s'])
```

```python
cricket_data.head()
```

```python
cricket_data.dtypes
```

```python
cricket_data["Matches"].unique()
```

```python
cricket_data["High_Score"].unique()
```

```python
cricket_data["Balls_Faced"].unique()
```

```python
cricket_data["Strike_Rate"].unique()
```

```python
cricket_data[["Matches", "High_Score", "Balls_Faced"]] = cricket_data[["Matches", "High_Score", "Balls_Faced"]].replace("[*+]", "", regex = True).astype({"Matches" : "int", "High_Score": "int"})
```

```python
cricket_data.dtypes
```

```python
cricket_data[["Balls_Faced", "Strike_Rate"]] = cricket_data[["Balls_Faced", "Strike_Rate"]].replace("-",np.nan, regex = True).astype(float)
```

```python
cricket_data.head()
```

```python
cricket_data.dtypes
```

```python
cricket_data.isnull().sum()
```

```python
cricket_data[cricket_data["Balls_Faced"].isna()]
```

```python
mean_for_8_14 = cricket_data[(cricket_data["Innings"].isin(range(60, 90))) & (cricket_data["Balls_Faced"].notna()) & cricket_data["Current/End_Year"].isin(range(1950,1970))]["Strike_Rate"].mean()
```

```python
cricket_data.loc[[8,14], "Strike_Rate"] = mean_for_8_14
```

```python
in_54 = cricket_data[cricket_data["Innings"].isin(range(25,40)) & cricket_data["Runs"].isin(range(1200,1600)) & cricket_data["Balls_Faced"].notna() & cricket_data["Current/End_Year"].isin(range(1900,1930))]["Strike_Rate"]
```

```python
cricket_data.loc[54, "Strike_Rate"] = 44.86
```

```python
cricket_data[cricket_data["Balls_Faced"].isna()]
```

```python
cricket_data.loc[[8,14,54], "Balls_Faced"] = round((cricket_data.loc[[8,14,54], "Runs"] * 100.00) / cricket_data.loc[[8,14,54], "Strike_Rate"]).astype(int)
```

#### Analysis by Country

```python
cricket_data.groupby("Country").agg(Total_Runs = ("Runs", "sum"), Average_Strike_Rate = ("Strike_Rate", "mean"), Total_Centuries = ("100s", "sum"), Total_Half_Centuries = ("50s", "sum")).sort_values("Total_Runs", ascending = False)
```

#### Top 5 and bottom 5 Payers by runs

```python
cricket_data.nlargest(5, "Runs")[["Country","Player", "Runs"]]
```

```python
cricket_data.nsmallest(5, "Runs")[["Country","Player", "Runs"]]
```

#### Top scoring player of every country

```python
cricket_data[cricket_data.groupby('Country')['Runs'].transform('max') == cricket_data['Runs']].sort_values("Runs", ascending = False)[["Country","Player", "Runs"]]
```

#### Highest Batting Average of a single player from every country

```python
cricket_data["Batting_Average"] = round(cricket_data["Runs"] / cricket_data["Innings"],2) 
```
